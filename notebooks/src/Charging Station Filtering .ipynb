```python
# Filtering 
```

```python
# total no. of places 
de1['State Name'].value_counts().sum()
```

```python
# are they unique
de1['State Name'].nunique()
```

```python
# count the 2 , 3, 4 wheel in diff state 
de1.groupby('State Name')[['Two Wheeler', 'Three Wheeler','Four Wheeler']].value_counts()
```

```python
# determine the count of domestic and public based  electric vehicle 
de1.groupby('State Name')[['Goods Vehicles', 'Public Service Vehicle','Special Category Vehicles', 'Ambulance/Hearses']].value_counts()
```

```python
# Which state has the highest number of charging stations?
de1[['State Name', 'total-charging-stations']].sort_values(by='total-charging-stations', ascending=False)
```

```python
# What percentage of the nation’s total charging stations does each state hold?
de1['Charging_Station_Share (%)'] = (de1['total-charging-stations'] / de1['total-charging-stations'].sum()) * 100
```

```python
# What is the distribution of total registered vehicles across states?
de1 = de1.sort_values(by="Grand Total", ascending=False)

# Plot
plt.figure(figsize=(8,8))
plt.barh(de1["State Name"], de1["Grand Total"], color='teal')
plt.xlabel("Total Registered Vehicles")
plt.title("Total Registered Electric Vehicles by State")
plt.gca().invert_yaxis()
plt.show()

```

```python
# Which states have zero charging stations?
de1[de1['total-charging-stations'].isnull() | (de1['total-charging-stations'] == 0)]
```

```python
# Which states have many vehicles but very few charging stations?
de1[(de1['Grand Total'] > 10000) & (de1['total-charging-stations'] < 20)]
```

```python
# Which states have more than 50,000 electric two-wheelers?
de1[de1['Two Wheeler'] > 50000]
```

```python
# count of vehicle used in construction site 
de1.groupby('State Name')['Construction Equipment Vehicle'].value_counts()
```

```python
de1.groupby('State Name')['Other'].value_counts()
```

```python
# value count of charging station per place 
de1.groupby('State Name')['total-charging-stations'].value_counts()
```

```python
# States where the number of charging stations is less than 500 (indicating potential infrastructure gaps)
de1[de1['total-charging-stations'] < 500]
```

```python
de1.groupby('State Name')[['Two Wheeler', 'Three Wheeler','Four Wheeler']].sum().sum(axis =1)
```

```python
# WHICH TYPE OF ELECTRIC VEHICLE IS MOST POPULAR IN INDIA 
de1.groupby('State Name')[['Two Wheeler', 'Three Wheeler','Four Wheeler']].sum().sum(axis =0).sort_values(ascending = False)
```

```python
# Which state has the highest number of electric vehicles overall?
de1[['State Name', 'Grand Total']].sort_values(by='Grand Total', ascending=False)

```

```python
# Which type of electric vehicle (2W, 3W, 4W) is most popular across India?
de1.groupby('State Name')[['Two Wheeler', 'Three Wheeler','Four Wheeler']].sum().sum(axis =0).sort_values(ascending = False)
```

```python
# Which states have outlier values—either very high or very low counts—for specific categories (e.g., 'Goods Vehicles', 'Special Category Vehicles')?
# By state for goods vehicles
q1 = de1['Goods Vehicles'].quantile(0.25)
q3 = de1['Goods Vehicles'].quantile(0.75)
iqr = q3 - q1
outliers = de1[(de1['Goods Vehicles'] < (q1 - 1.5*iqr)) | (de1['Goods Vehicles'] > (q3 + 1.5*iqr))]
print(outliers[['State Name', 'Goods Vehicles']])
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.boxplot(de1['Goods Vehicles'], vert=False)
plt.title('Distribution of Electric Goods Vehicles by State (2025)')
plt.xlabel('Number of Goods Vehicles')
plt.show()
```

```python
##Which state has the highest number of two-wheeler EVs?
de1[['State Name', 'Two Wheeler']].sort_values(by='Two Wheeler', ascending=False)
```

```python
# Which states have the highest number of EV charging stations?

top_states = de1.sort_values(by='total-charging-stations', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top_states['State Name'], top_states['total-charging-stations'], color='royalblue')
plt.title('Top 10 States by Number of EV Charging Stations (2025)')
plt.xlabel('Number of Charging Stations')
plt.ylabel('State')
plt.gca().invert_yaxis()  # Largest bar at the top
plt.tight_layout()
plt.show()
```

```python
# Which states have EVs but no charging stations?
de1[de1['total-charging-stations'].isna() & (de1['Grand Total'] > 0)]
```

```python
import matplotlib.pyplot as plt

# Filter the states with EVs but no charging stations
no_charging_states = de1[de1['total-charging-stations'].isna() & (de1['Grand Total'] > 0)]

# Sort values for better visual clarity
no_charging_states_sorted = no_charging_states.sort_values(by='Grand Total', ascending=True)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(no_charging_states_sorted['State Name'], no_charging_states_sorted['Grand Total'], color='orange')
plt.xlabel('Number of EVs (Grand Total)')
plt.title('States with EVs but No Charging Stations')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# How does the number of charging stations relate to total EVs per state?
(de1['total-charging-stations']).corr(de1['Grand Total'] )
```

```python
# Are certain states specializing in specific types of vehicles but lacking adequate chargers?
# Example: States with >50,000 Three Wheelers and <500 chargers
specialized = de1[(de1['Three Wheeler'] > 50000) & (de1['total-charging-stations'] < 500)]
print(specialized[['State Name', 'Three Wheeler', 'total-charging-stations']])
```

```python
# What is the average number of charging stations per 1,000 EVs in each state?
de1['Chargers per 1k EVs'] = (de1['total-charging-stations'] / de1['Grand Total']) * 1000
print(de1[['State Name', 'Chargers per 1k EVs']].sort_values(by='Chargers per 1k EVs', ascending=False))
# bar chart to show this
de1_sorted = de1.sort_values(by='Chargers per 1k EVs', ascending=False)
plt.figure(figsize=(10,6))
plt.barh(de1_sorted['State Name'], de1_sorted['Chargers per 1k EVs'], color='teal')
plt.xlabel('Charging Stations per 1,000 EVs')
plt.title('Charger Density by State (2025)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# What is the EV-to-charging-station ratio for each state?
de1['EV_per_Station'] = (de1['Grand Total'] / de1['total-charging-stations'])
```

```python
#Which states need urgent installation of charging stations?
de1[de1['total-charging-stations'] < 10].sort_values(by='Grand Total', ascending=False)
```

```python
# Which states might benefit from boosting 4-wheeler EV infrastructure?
de1[(de1['Two Wheeler'] > 5000) & (de1['Four Wheeler'] < 100)]
```

```python
# Function to removce outlier from multiple columns
def remove_outliers_iqr_multi(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
        
    for col in columns:
        sns.boxplot(y=df_clean[col])
        plt.title(f'Boxplot of {col} (Outliers Removed)')
        plt.show()
    return df_clean
```

```python
num = charge.select_dtypes(exclude = 'object').columns 
charge = remove_outliers_iqr_multi(charge, num)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
charge['Charging_Station_Required'] = (
    charge['EV Registrations (in thousands)'] * 1000 / 150 +
    charge['State Population (millions)'] * 0.5 +
    charge['Smart Cities Count'] * 2
)
```

```python
charge_filtered = charge.drop(columns=[
    'State Name',                       # Categorical, optional for spatial plots only
    'Charging_Station_Share (%)',      # Derived from station count
    'EV_per_Station'                   # Also derived
])
```

```python
X = charge_filtered.drop(columns='Charging_Station_Required')
y = charge_filtered['Charging_Station_Required']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
#Make Predictions
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#10. Evaluate Model
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Root Mean Squared Error (RMSE):", rmse)
print("R² Score:", r2)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', edgecolor='k')
plt.xlabel("Actual Charging Station Requirement")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Charging Station Requirement")
plt.grid(True)
plt.show()
```

```python
importances = model.feature_importances_
features = X.columns

plt.figure(figsize=(12, 6))
sns.barplot(x=importances, y=features)
plt.title("Feature Importances")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
charge['Predicted_Stations'] = model.predict(scaler.transform(X))
result = charge[['State Name', 'EV Registrations (in thousands)', 'Predicted_Stations']]
print(result.sort_values(by='Predicted_Stations', ascending=False))
```
